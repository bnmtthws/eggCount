```r
library(ggplot2)
library(plyr)
library(reshape2)
```

```r
eggData_20160920 <- read.csv('~/Dropbox/stephanie marcus rotation//eggCounts//20160920/pixel_counts.csv')

eggData_20160920 
```

```r
eggData_20160920.filtered <- eggData_20160920[eggData_20160920$total.pixel > 150,]

preference_plot.filtered.jitter <- ggplot(data=eggData_20160920.filtered,aes(x=genotype,y=preference.index,colour=total.pixel)) + geom_jitter(width=0.5,alpha=0.5)
preference_plot.filtered.violin <- ggplot(data=eggData_20160920.filtered,aes(x=genotype,y=preference.index,colour=total.pixel)) + geom_violin() + geom_jitter(width=0.5,alpha=0.5)
```
